#Basic Reshape! üîÑ

#Auto-dimension with -1! ü§ñ

````python
arr = np.arange(1, 13)
print("Original array:")
print(arr)
print("Shape:", arr.shape)
````

````python
print("Reshape (3, -1) - auto calculates 4:")
r1 = arr.reshape(3, -1)
print(r1)
print("Shape:", r1.shape)
````

````python
print("Reshape (-1, 4) - auto calculates 3:")
r2 = arr.reshape(-1, 4)
print(r2)
print("Shape:", r2.shape)
````

````python
print("Reshape (-1, 1) - column vector:")
r3 = arr.reshape(-1, 1)
print(r3)
print("Shape:", r3.shape)
````

````python
print("Reshape (1, -1) - row vector:")
r4 = arr.reshape(1, -1)
print(r4)
print("Shape:", r4.shape)
````

#Ravel vs Flatten! üìè

````python
arr = np.array([[1, 2, 3],
                [4, 5, 6]])

print("Original 2D array:")
print(arr)
print("Shape:", arr.shape)
````

````python
print("RAVEL (view):")
raveled = arr.ravel()
print(raveled)
print("Shape:", raveled.shape)
````

Test if it's a view

````python
print("Modify raveled[0] = 999:")
raveled[0] = 999
print("Raveled:", raveled)
print("Original arr:", arr)
print("‚Üí Original changed! (view)")
````

Reset

````python
arr[0, 0] = 1
````

FLATTEN - returns copy

````python
print("FLATTEN (copy):")
flattened = arr.flatten()
print(flattened)
print("Shape:", flattened.shape)
````

Test if it's a copy

````python
print("Modify flattened[0] = 999:")
flattened[0] = 999
print("Flattened:", flattened)
print("Original arr:", arr)
print("‚Üí Original not changed! (copy)")
````

#Real Example - Image Data! üñºÔ∏è

````python
np.random.seed(42)
image = np.random.randint(0, 256, (10, 10))
````

````python
print("Original image (10, 10):")
print(image)
print("Shape:", image.shape)
````

Flatten for processing

````python
print("Flatten to 1D for ML model:")
flat = image.flatten()
print(f"Flattened shape: {flat.shape}")
print(f'First 10 pixels: {flat[:10]}')
````

Reshape back to image

````python
print("Reshape back to 2D:")
restored = flat.reshape(10, 10)
print(restored)
print("Shape:", restored.shape)
````

Verify

````python
print("Images match:", np.array_equal(image, restored))
````

Reshape to different size (if needed)

````python
print("Reshape to 5x20:")
reshaped = flat.reshape(5, 20)
print(reshaped)
print("Shape:", reshaped.shape)
````

````python
r = reshaped.reshape(20, 5)
print(r)
````

#Flatten for Analysis

````python
np.random.seed(42)
weekly_temperature = np.random.randint(15, 30, (4, 7))
print("Shape:", weekly_temperature.shape)
print(weekly_temperature)
````

````python
flattened = weekly_temperature.flatten()
print(flattened)
print("Shape:", flattened.shape)
````

````python
overal_average = flattened.mean()
print(f"Average temperature: {overal_average:.1f}")
````

````python
overal_min = flattened.min()
print(f"Minimum temperature: {overal_min}")
````

````python
overal_max = flattened.max()
print(f"Maxsimum temperature: {overal_max}")
````

````python
weekly_avg = weekly_temperature.mean(axis=1)
print(f"Weekly average temperature: {weekly_avg}")
````

````python
dayly_avg = weekly_temperature.mean(axis=0)
print(f"Dayly average temperature: {dayly_avg}")
````
