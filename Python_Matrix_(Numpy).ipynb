```python
def can_multiply(mat1,mat2):
  if mat1.shape[1]==mat2.shape[0]:
    result_size=(mat1.shape[0],mat2.shape[1])
    print(f"These Matrics can be multiplied with the result size being {result_size}")
  else:
    print( "These Matrics cannot be multiplied")
```
