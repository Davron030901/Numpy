#Weekly sales

````python
sales = np.array([[100, 120, 110, 130, 150],
                  [105, 115, 115, 125, 145],
                  [110, 125, 120, 135, 155]])
print("Weekly sales:")
print(sales)
````

Average daily sales (axis=0)

````python
daily_avg = sales.mean(axis=0)
days = ["Monday", "Tuesday", "Wednesday", "Thursday", "Friday"]
print("Average daily sales:")
for day, avg in zip(days, daily_avg):
  print(f"{day}: ${avg:.1f}")
````

#Class grade ðŸ“š

````python
grades = np.array([[85, 78, 92],
                   [90, 82, 88],
                   [78, 85, 95],
                   [88, 80, 90]])
print("Class grades:")
print(grades)
````

Average grade for each subject (axis=0)

````python
subject_avg = grades.mean(axis=0)
subjects = ["Math", "Physics", "English"]

print("Average grade for each subject:")
for subject, avg in zip(subjects, subject_avg):
  print(f"{subject}: {avg:.1f}")
````

#Monthly temperature

````python
temp = np.array([[-5,    0,     8,     15],
                 [-10,   -5,    5,     12],
                 [-3,    2,     10,    18]])

print("Monthly temperatures:")
print(temp)
````

Average temperature of each MONTH (axis=0)

````python
avg_temp = temp.mean(axis=0)
months = ["January", "February", "March", "April"]
print("Average temperature of each MONTH:")
for month, avg in zip(months, avg_temp):
  print(f"{month}: {avg:.1f}")
````

#Store sales

````python
sales = np.array([[50,   30,    20],
                  [45,   35,    25],
                  [55,   28,    22],
                  [48,   32,    18]])
print("Store sales:")
print(sales)
````

Total sales of each PRODUCT (axis=0)

Total sales of each PRODUCT (axis=0)

````python
total_sales = sales.sum(axis=0)
print("Total sales of each PRODUCT:")
product = ["Tea", "Sugar", "Oil"]
for prod, total in zip(product, total_sales):
  print(f"{prod}: {total}")
````

Average sales of each PRODUCT (axis=0)

````python
avg_sales = sales.mean(axis=0)
market = ["Market1", "Market2", "Market3"]
print("Average sales of each PRODUCT:")
for market, avg in zip(market, avg_sales):
  print(f"{market}: {avg:.1f}")
````

#Sports team

````python
points = np.array([[12,   15,    18],
                   [10,   12,    14],
                   [15,   18,    20],
                   [8,    10,    12],
                   [14,   16,    19]])
print("Sports team points:")
print(points)
````

Total points per GAME (axis=0)

````python
total_points = points.sum(axis=0)
games = ["Game1", "Game2", "Game3"]
print("Total points per GAME:")
for game, total in zip(games, total_points):
  print(f"{game}: {total}")
````

Average points per GAME (axis=0)

````python
avg_points = points.mean(axis=0)
print("Average points per GAME:")
for game, avg in zip(games, avg_points):
  print(f"{game}: {avg:.1f}")
````

Which game scored the most points?

````python
max = points.max(axis=0)
print("Which game scored the most points?")
for game, max in zip(games, max):
  print(f"{game}: {max}")
````
