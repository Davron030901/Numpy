```python
import numpy as np
```

#Set + String Operations

```python
email_list_a = np.array(['alice@gmail.com', 'bob@yahoo.com', 'charlie@gmail.com'], dtype="U30")
email_list_b = np.array(['bob@yahoo.com', 'diana@gmail.com', 'alice@gmail.com'], dtype="U30")

print("Email list A:", email_list_a)
print("Email list B:", email_list_b)
```

```python
both_intersected = np.intersect1d(email_list_a, email_list_b)
print("Both lists (intersect)")
print(both_intersected)
```

```python
users_union = np.union1d(email_list_a, email_list_b)
print("Users (union)")
print(users_union)
```

```python
only_list_a = np.setdiff1d(email_list_a, email_list_b)
print("Only list A")
print(only_list_a)
```

```python
uppercase_all_emails = np.char.upper(users_union)
print("Uppercase all emails")
print(uppercase_all_emails)
```

#Linear Algebra + Statistics

```python
sales_matrix = np.array([[100, 120, 130, 140],
                         [200, 180, 190, 210],
                         [150, 160, 155, 165]])
print("Sales matrix:")
print(sales_matrix)
```

```python
weights = np.array([[0.2], [0.3], [0.3], [0.2]])
print("Weights:", weights)
```

```python
matrix_mult = sales_matrix @ weights
print("Matrix multiplication:")
print(matrix_mult)
```

```python
q1 = np.percentile(matrix_mult, 25)
q3 = np.percentile(matrix_mult, 75)
print("Q1:", q1)
print("Q2:", q3)
```

```python
correlation_matrix = np.corrcoef(sales_matrix)
print("Correlation matrix:")
print(correlation_matrix)
```

```python
correlation_product1_product2 = correlation_matrix[0, 1]
print(f"Correlation between Product 1 and Product 2: {correlation_product1_product2:.4f}")
```

#String + Date Combined

```python
events = np.array(['Conference', 'Workshop', 'Seminar', 'Meeting'])
dates = np.array(['2024-03-15', '2024-02-20', '2024-04-10', '2024-03-01'], dtype='datetime64')

print("Events:", events)
print("Dates:", dates)
```

```python
find_events_with_e = np.char.find(events, 'e')
print("Events with 'e':", find_events_with_e)
find_0 = find_events_with_e >= 0
print("Events with 'e' in their name:", events[find_0])
```

```python
sort_dates = np.sort(dates)
print("Sorted dates:", sort_dates)
```

```python
event_in_march = sort_dates.astype('datetime64[M]') == np.datetime64('2024-03')
print("Events in March:", events[event_in_march])
```

```python
events_sufix = np.char.add(events, " 2024")
print("Events 2024:", events_sufix)
```

#Statistics + Outliers

```python
respons_time = np.array([120, 135, 145, 150, 155, 160, 165,
                      170, 180, 500, 140, 148, 152])

print("Response time (ms):", respons_time)
```

```python
q1 = np.percentile(respons_time, 25)
q3 = np.percentile(respons_time, 75)
median = np.median(respons_time)
print("Q1:", q1)
print("Q3:", q3)
print("Median:", median)
```

```python
iqr = q3 - q1
print("IQR:", iqr)
```

```python
outlier = (respons_time > (q3 + 1.5 * iqr)) | (respons_time < (q1 - 1.5 * iqr))
outlier_tukeys = respons_time[outlier]
print("Outliers (Tukey's method):", outlier_tukeys)
```

```python
clean_data = respons_time[~outlier]
print("Clean data:", clean_data)
```

#Complete Integration

```python
names = np.array(['Alice Smith', 'Bob Johnson', 'Charlie Brown'])
join_dates = np.array(['2020-01-15', '2019-06-20', '2021-03-10'], dtype='datetime64')
salaries = np.array([[3000, 3200, 3400, 3600],
            [4000, 4200, 4400, 4600],
            [2800, 3000, 3100, 3300]])

print("Names:", names)
print("Join dates:", join_dates)
print("Salaries:")
print(salaries)
```

```python
today = np.datetime64('2024-02-22')
print("Today:", today)
```

```python
years_of_service = (today - join_dates) / 365
print("Years of service:", years_of_service)
```

```python
total = salaries.sum(axis=1)
print("Total salaries:", total)
```

```python
seniaor_employees = years_of_service > 3
print("Senior employees:", names[seniaor_employees])
```

```python
print("Create summary:")
for name, service, total_salary in zip(names, years_of_service, total):
    print(f"Name:{name}, sevice: {service}, total:${total_salary}")
```
