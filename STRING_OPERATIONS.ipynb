````python
import numpy as np
````

#Case Operations! 🔤

````python
names = np.array(['Alice', 'BOB', 'Charlie', 'DIANA'], dtype='U20')
print("Original names:")
print(names)
````

UPPERCASE

````python
upper = np.char.upper(names)
print("Uppercase names:")
print(upper)
````

 LOWERCASE

````python
lower = np.char.lower(names)
print("Lowercase names:")
print(lower)
````

CAPITALIZE (first letter upper)

````python
capitalized = np.char.capitalize(names)
print("Capitalized names:")
print(capitalized)
````

TITLE CASE

````python
text = np.array(['hello world', 'python numpy'], dtype='U20')
title = np.char.title(text)
print("Title case:")
print(title)
````

#String Info & Search! 🔍

````python
emails = np.array(['alice@gmail.com', 'bob@yahoo.com',
                   'charlie@gmail.com', 'diana@outlook.com'], dtype='U30')
print("Emails:")
print(emails)
````

STRING LENGTH

````python
lengths = np.char.str_len(emails)
print("Email lengths:")
print(lengths)
````

FIND substring (returns index, -1 if not found)

````python
gmail_indices = np.char.find(emails, 'gmail')
print("Find 'gmail' (index):")
print(gmail_indices)
````

Boolean mask: contains 'gmail'

````python
is_gmail = gmail_indices >= 0
print("Boolean mask for 'gmail':")
print(is_gmail)
````

Filter Gmail users

````python
gmail_users = emails[is_gmail]
print("Gmail users:")
print(gmail_users)
````

STARTSWITH

````python
starts_with_a = np.char.startswith(emails, 'a')
print("Starts with 'a':")
print(starts_with_a)
````

ENDSWITH

````python
ends_with_com = np.char.endswith(emails, '.com')
print("Ends with '.com':")
print(ends_with_com)
````

#String Manipulation! ✂️

````python
codes = np.array(['ABC-001', 'ABC-002', 'XYZ-003', 'XYZ-004'], dtype='U10')
print("Original codes:")
print(codes)
````

REPLACE

````python
new_codes = np.char.replace(codes, 'ABC', 'DEF')
print("Replaced codes:")
print(new_codes)
````

ADD (concatenate)

````python
names = np.array(['Alice', 'Bob', 'Charlie'], dtype='U20')
last_name = ' Smith'
full_names = np.char.add(names, last_name)
print("Full names:")
print(full_names)
````

STRIP (remove whitespace)

````python
messy = np.array(['  hello  ', 'world  ', '  numpy'], dtype='U20')
cleaned = np.char.strip(messy)
print("Strip whitespace:")
print(f"Before: {messy}")
print(f"After: {cleaned}")
````

MULTIPLY (repeat)

````python
symbols = np.array(['*', '-', '+'], dtype="U5")
repeated = np.char.multiply(symbols, 5)
print("Repeated symbols:")
print(repeated)
````

#Real Example - Name Processing! 👤

````python
raw_names = np.array(['  ALICE JOHNSON  ', 'bob SMITH',
                      'Charlie  BROWN', '  diana DAVIS  '], dtype='U30')
print("Raw names:")
for i, name in enumerate(raw_names):
    print(f"{i+1}. {name}")
````

Strip whitespace

````python
step1 = np.char.strip(raw_names)
print("Step 1 - Strip:")
print(step1)
````

Title case

````python
step2 = np.char.title(step1)
print("Step 2 - Title case:")
print(step2)
````

Get first names

````python
first_names = np.array([name.split()[0] for name in step2])
print("Step 3 First names:")
print(first_names)
````

Get initials (first letter)

````python
initials = np.char.upper(np.char.array([name[0] for name in first_names]))
print("Step 4 - Initials:")
print(initials)
````

Final summary

````python
print("Summary:")
for raw, clean, first, initial in zip(raw_names, step2, first_names, initials):
  print(f"'{raw.strip()}' →  {clean} → {first} ({initial})")
````

#Product Code Formatting

#Name Analysis

````python
names = np.array(['Alice Smith', 'Bob Johnson', 'Charlie Brown',
        'Diana Wilson', 'Eve Davis'], dtype='U20')

print("Names:")
print(names)
````

````python
each_name_leng = np.char.str_len(names)
print("Length of each name:")
print(each_name_leng)
````

````python
best_name =np.argmax(each_name_leng)
print(f"Best name:{names[best_name]}")
````

````python
lower = np.char.lower(names)
print("Lowercase names:")
print(lower)
````
