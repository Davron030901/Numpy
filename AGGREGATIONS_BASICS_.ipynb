#Monthly expenses

```python
expenses = np.array([150, 200, 180, 170])
print("Monthly expenses:", expenses)
```

Monthly total

```python
monthly_total = expenses.sum()
print("Monthly total:", monthly_total)
```

Weekly average

```python
weekly_avg = expenses.mean()
print("Weekly average:", weekly_avg)
```

Budget analysis

```python
budget = 800
print("Budget:", budget)
```

```python
if monthly_total > budget:
  saved = budget - monthly_total
  print(f"Budget saved {saved}")
else:
  over = monthly_total - budget
  print(f"Budget over by {over}")
```

#Transportation cost (sum)

```python
transport = np.array([5000, 5000, 6000, 5500, 5000,
                      4500, 5000, 5500, 6000, 5000,
                      5000, 5500, 5000, 4500, 5000,
                      5500, 6000, 5000, 5000, 5500,
                      5000, 4500, 5000, 5500, 6000,
                      5000, 5000, 5500, 5000, 4500])
print("Transportation cost:", transport)
```

Monthly total sum

```python
total_monthly = transport.sum()
print("Monthly total sum:", total_monthly)
```

Monthly average

```python
average_monthly = transport.mean()
print(f"Monthly average:{average_monthly:.1f}")
```
