````python
import numpy as np
import os
````

#np.savez() - Multiple Arrays! 📦

````python
student_ids = np.array([101, 102, 103, 104, 105])
scores = np.array([[85, 90, 78],
                   [92, 88, 95],
                   [78, 85, 82],
                   [88, 92, 90],
                   [95, 93, 96]])
subjects = np.array(['Math', 'Physics', 'English'], dtype='U10')

print("Student IDs:", student_ids)
print("Scores:")
print(scores)
print("Subjects:", subjects)
````

SAVE all together

````python
np.savez("student_data.npz",
         ids=student_ids,
         scores=scores,
         subjects=subjects)
print("Saved to 'student_data.npz'")
````

LOAD all

````python
print("Loading data ...")
data = np.load("student_data.npz")
print("Available keys:", list(data.keys()))
````

Access by key

````python
loaded_ids = data['ids']
loaded_scores = data['scores']
loaded_subjects = data['subjects']

print("Loaded Student IDs:", loaded_ids)
print("Loaded Scores:")
print(loaded_scores)
print("Loaded Subjects:", loaded_subjects)
````

Verify

````python
print("IDs match:", np.array_equal(student_ids, loaded_ids))
print("Scores match:", np.array_equal(scores, loaded_scores))
print("Subjects match:", np.array_equal(subjects, loaded_subjects))
````

Close file

````python
data.close()
````

#Compressed Format! 🗜️

````python
large_data = np.random.rand(1000, 1000)
print(f"Large array shape: {large_data.shape}")
````

SAVE uncompressed

````python
np.savez("large_uncompressed.npz", data=large_data)
size_uncompressed = os.path.getsize("large_uncompressed.npz")
print(f"Size of uncompressed file: {size_uncompressed / 1024:.2f} KB")
print("Saved to 'large_uncompressed.npz'")
````

SAVE compressed

````python
np.savez_compressed('large_compressed.npz', data=large_data)
size_compressed = os.path.getsize('large_compressed.npz')
print(f"Size of compressed file: {size_compressed / 1024:.2f} KB")
````

Compression ratio

````python
ratio = size_uncompressed / size_compressed
print(f"Compression ratio: {ratio:.2f}x smaller")
````

Load compressed (same way)

````python
data = np.load("large_compressed.npz")
loaded = data['data']
print("Loaded shape:", loaded.shape)
print("Data match:", np.array_equal(large_data, loaded))
````

#Binary Files! ⚡

````python
data = np.array([1.5, 2.7, 3.9, 4.1, 5.3])
print("Original data:", data)
print(f"dtype: {data.dtype}, shape: {data.shape}")
````

````python
data.tofile('data.bin')
print("Saved ti 'data.bin' (binary)")
````

LOAD from binary

IMPORTANT: Must specify dtype and shape!

````python
loaded = np.fromfile('data.bin', dtype=np.float64)
print("Loaded data: ", loaded)
print(f"dtype: {loaded.dtype}, shape: {loaded.shape}")
````

Verify

````python
print("Data match:", np.array_equal(data, loaded))
````

````python
matrix = np.array([[1, 2, 3],
                   [4, 5, 6]], dtype=np.int32)

print("Matrix:")
print(matrix)
````

````python
matrix.tofile('matrix.bin')
````

````python
loaded_flat = np.fromfile('matrix.bin', dtype=np.int32)
loaded_matrix = loaded_flat.reshape(2, 3)
print("Loaded flat array:", loaded_flat)
print("Loaded matrix:")
print(loaded_matrix)
````

#Default Names in savez! 📝

````python
arr1 = np.array([1, 2, 3])
arr2 = np.array([4, 5, 6])
arr3 = np.array([7, 8, 9])

print("Array 1:", arr1)
print("Array 2:", arr2)
print("Array 3:", arr3)
````

Save without keywords (uses arr_0, arr_1, ...)

````python
np.savez("default_namez.npz", arr1, arr2, arr3)
print("Saved to 'default_namez.npz'")
````

Load

````python
data = np.load("default_namez.npz")
print("Available keys:", list(data.keys()))
````

Access by default names

````python
print("arr_0:", data['arr_0'])
print("arr_1:", data['arr_1'])
print("arr_2:", data['arr_2'])
````

#Project Data Archive

````python
project_name = np.array('WebApp')
team_members = np.array(['Alice', 'Bob', 'Charlie', 'Diana'], dtype='U10')
sprint_scores = np.array([[85, 90, 88], [92, 87, 91], [78, 82, 85], [95, 93, 96]])
sprint_dates = np.array(['2024-01-01', '2024-02-01', '2024-03-01'], dtype='datetime64')

print("Project Name:", project_name)
print("Team Members:", team_members)
print("Sprint Scores:")
print(sprint_scores)
````

````python
np.savez("project.npz",
         name = project_name,
         members = team_members,
         scores = sprint_scores,
         dates = sprint_dates)
print("Saved to 'project.npz'")
````

````python
data = np.load("project.npz")
print("Available keys:", list(data.keys()))
````

````python
print("Project name:", data['name'])
print("Team members:", data['members'])
print("Sprint scores:")
print(data['scores'])
print("Total sprints:", data['dates'])
````

Calculate average score per sprint

````python
avg_sprint = data['scores'].mean(axis=0)
for sprint, avg_scores in zip(data['dates'], avg_sprint):
    print(f"Average scores for sprint {sprint}: {avg_scores}")
````

#Binary Format Speed

````python
response_times = np.array([120, 135, 145, 150, 155, 160, 165, 170], dtype=np.int32)
timestamps = np.array([0, 5, 10, 15, 20, 25, 30, 35], dtype=np.int32)

print("Response times:", response_times)
print("Timestamps:", timestamps)
````

````python
response_times.tofile("times.bin")
timestamps.tofile("stamps.bin")
````

````python
times = np.fromfile("times.bin", dtype=np.int32)
stamps = np.fromfile("stamps.bin", dtype=np.int32)

print("Loaded response times:", times)
print("Loaded timestamps:", stamps)
````

Verify data integrity

````python
print("Verify data integrity times:", np.array_equal(response_times, times))
print("Verify data integrity stamps:", np.array_equal(timestamps, stamps))
````
