# NUMPY SPECIAL ARRAYS

```python
import numpy as np
```

5 zeros

```python
zeros_1d=np.zeros(5)
print(zeros_1d)
```

3x3 zeros (2D matrix)

```python
zeros_2d=np.zeros((3,3))
print(zeros_2d)
```

2x4 zeros (2 rows, 4 columns)

```python
zeros_2x4=np.zeros((2,4))
print(zeros_2x4)
```

4 units (1D)

```python
ones_1d=np.ones(4)
print(ones_1d)
```

2x2 units (2D)

```python
ones_2d=np.ones((2,2))
print(ones_2d)
```

3x5 units

```python
ones_3x5=np.ones((3,5))
print(ones_3x5)
```

**Simulate an image**

"Image" = matrix 10x10 black image (pixels = 0)

```python
black_image=np.zeros((10,10))
print(f"Black image:\n {black_image}")
```

10x10 white image (pixels = 1)

```python
white_image=np.ones((10,10))
print(f"White image: \n {white_image}")
```

Shape check

```python
print(f"Black image shape: {black_image.shape}")
print(f"White image shape: {white_image.shape}")
```

7 zeros (1D)

```python
zero_1d=np.zeros(7)
print(zero_1d)
```

4x4 matrix of zeros

```python
zeros_4x4=np.zeros((4,4))
print(zeros_4x4)
```

5x2 units (5 rows, 2 columns)

```python
ones_5x2=np.ones((5,2))
print(ones_5x2)
```

100x100 "black picture"

```python
black_100x100=np.zeros((100,100))
print(black_100x100)
```

50x50 "white painting"

```python
white_50x50=np.ones((50,50))
print(white_50x50)
```

Print 3x3 zeros and 3x3 ones side by side

```python
zeros_3x3=np.zeros((3,3))
ones_3x3=np.ones((3,3))
print(zeros_3x3,ones_3x3)
```
