#Adds each row (horizontal ‚Üí)

````python
grades = np.array([[85, 90, 78, 92],
                   [88, 85, 82, 90],
                   [90, 88, 85, 95]])
print("All students grades:")
print(grades)
````

````python
total_per_student = grades.sum(axis=1)
students = ["Ali", "Vali", "Guli"]
for student, total in zip(students, total_per_student):
  print(f"{student}: {total}")
````

#Student evaluation üìö

````python
grades = np.array([[85, 90, 78, 92, 88],
                   [88, 85, 82, 90, 86],
                   [90, 88, 85, 95, 92],
                   [82, 80, 88, 85, 83]])

students = ['Ali', 'Vali', 'Guli', 'Diyor']

print("All students grades:")
print(grades)
````

````python
student_avg = grades.mean(axis=1)

print("Student average:")
for student, avg in zip(students, student_avg):
  print(f"{student}: {avg}")
````

````python
print(f"{students[student_avg.argmax()]} has the best average with {student_avg.max()}")
````

Monthly expenses üí∞

````python
expenses = np.array([[500, 300, 150, 100, 80],
                     [600, 350, 200, 120, 100],
                     [450, 280, 130, 90, 70],
                     [550, 320, 180, 110, 90]])
families = ["Family 1", "Family 2", "Family 3", "Family 4"]

print("All expenses:")
print(expenses)
````

````python
total_per_family = expenses.sum(axis=1)

print("Total expenses per family:")
for family, total in zip(families, total_per_family):
  print(f"{family}: {total}")
````

````python
print(f"{families[total_per_family.argmax()]} has the most expenses with {total_per_family.max()}")
print(f"{families[total_per_family.argmin()]} has the least expenses with {total_per_family.min()}")
````

#Football team ‚öΩ

````python
goals = np.array([[2,    3,     1,     2],
                  [1,    2,     2,     3],
                  [3,    1,     2,     1],
                  [1,    1,     1,     2],
                  [2,    2,     3,     2]])
players = ["Ronaldo", "Messi", "Neymar", "Mbappe", "Haaland"]

print("All players goals:")
print(goals)
````

````python
total_goals_per_player = goals.sum(axis=1)

print("Total goals per player:")
for player, total in zip(players, total_goals_per_player):
  print(f"{player}: {total}")
````

````python
top_scorer_idx = total_goals_per_player.argmax()
print(f"{players[top_scorer_idx]} has the most goals with {total_goals_per_player[top_scorer_idx]}")
````

````python
avg_goals = goals.mean(axis=1)
print("Average goals per player:")
for player, avg in zip(players, avg_goals):
  print(f"{player}: {avg} goals/playing")
````

#Weekly working hours

````python
hours = np.array([[8,    9,    7,    8,    6],   # Aziz
                  [7,    8,    8,    9,    7],   # Botir
                  [9,    8,    8,    8,    8],   # Dilshod
                  [6,    7,    9,    7,    9]])  # Erkin

employees = ["Aziz", "Botir", "Dilshod", "Erkin"]

print("All employees hours:")
print(hours)
````

Total weekly hours per employee

````python
total_per_employee = hours.sum(axis=1)
for employee, total in zip(employees, total_per_employee):
  print(f"{employee}: {total} hours")
````

Average hours per day per employee

````python
avg_employee_per_day = hours.mean(axis=1)
for employee, avg in zip(employees, avg_employee_per_day):
  print(f"{employee}: {avg} hours/day")
````

````python
top_scorer_idx = total_per_employee.argmax()
print(f"{employees[top_scorer_idx]} has the most hours with {total_per_employee[top_scorer_idx]}")
````

#Store sales

````python
sales = np.array([[50,   30,    20],
                  [45,   35,    25],
                  [55,   28,    22],
                  [48,   32,    18],
                  [52,   33,    23]])
markets = ["Market 1", "Market 2", "Market 3", "Market 4", "Market 5"]

print("All markets sales:")
print(sales)
````

Total sales of each STORE

````python
total_per_markets = sales.sum(axis=1)
for market, total in zip(markets, total_per_markets):
  print(f"{market}: {total}")
````

Average sales per STORE

````python
avg_per_markets = sales.mean(axis=1)
for market, avg in zip(markets, avg_per_markets):
  print(f"{market}: {avg:.2f}")
````

````python
top_idx = total_per_markets.argmax()
print(f"{markets[top_idx]} has the most sales with {total_per_markets[top_idx]}")
````

#Test results

````python
scores = np.array([[75,   80,    85,    90],   # Student 1
                   [85,   82,    88,    92],   # Student 2
                   [70,   75,    78,    82],   # Student 3
                   [90,   88,    92,    95],   # Student 4
                   [65,   70,    72,    75],   # Student 5
                   [88,   85,    90,    93]])  # Student 6
students = ["Student 1", "Student 2", "Student 3", "Student 4", "Student 5", "Student 6"]

print("All students scores:")
print(scores)
````

Average score of each student

````python
avg_students = scores.mean(axis=1)
for student, avg in zip(students, avg_students):
  print(f"{student}: {avg:.2f}")
````

````python
top_idx = avg_students.argmax()
print(f"{students[top_idx]} has the best average with {avg_students[top_idx]}")
````

````python
low_idx = avg_students.argmin()
print(f"{students[low_idx]} has the worst average with {avg_students[low_idx]}")
````
