#Seed Magic! üå±

Seedless - different every time

```python
print("Run 1:", np.random.rand(3))
print("Run 2:", np.random.rand(3))
print("Run 3:", np.random.rand(3))
```

With Seed - the same every time!

```python
np.random.seed(32)
print("Run 1:", np.random.rand(3))
```

```python
np.random.seed(32)
print("Run 2:", np.random.rand(3))
```

```python
np.random.seed(32)
print("Run 3:", np.random.rand(3))
```

#Lottery - Try your luck! üé∞

Choose 6 numbers from 1-49

 Different result every time

```python
print("Your luck:")
for i in range(3):
  random_floats = np.random.rand(6)
  lottery_numbers = (random_floats * 49 + 1).astype(int)
  print(f"Attempt {i+1}: {lottery_numbers}")
```

#Exam Simulator üìù

```python
np.random.seed(100)
random_scores = np.random.rand(10)
scores = (random_scores * 100).astype(int)
print("Exam Scores:")
print(scores)
```

Statistics

```python
print(f"Average Score: {scores.mean():.1f}")
print(f"Highest Score: {scores.max()}")
print(f"Lowest Score: {scores.min()}")
```
