````python
import numpy as np
import matplotlib.pyplot as plt
````

#Multiple Lines - Sin & Cos! ðŸŒŠ

````python
x = np.linspace(0, 2*np.pi, 100)
y1 = np.sin(x)
y2 = np.cos(x)
y3 = y1 * y2
````

````python
plt.plot(x, y1, label="sin(x)", color="blue", linewidth=2)
plt.plot(x, y2, label="cos(x)", color='red', linewidth=2)
plt.plot(x, y3, label='sin*cos', color='green', linewidth=2, linestyle="--")

plt.title("Trigonometric Functions", fontsize=16)
plt.xlabel("X", fontsize=12)
plt.ylabel("Y", fontsize=12)
plt.legend(fontsize=10)
plt.grid(True, alpha=0.3)

plt.show()
````

#Real Data - Weekly Temperature! ðŸŒ¡ï¸

````python
days = np.array([1, 2, 3, 4, 5, 6, 7])
tashkent = np.array([20, 22, 25, 23, 21, 24, 26])
samarqand = np.array([18, 20, 22, 21, 19, 22, 24])
````

````python
plt.plot(days, tashkent, label="Tashkent", color='red', linewidth=2, marker='o')
plt.plot(days, samarqand, label='Samarqand', color='blue', linewidth=2, marker='s')

plt.title("Weekly Temperature", fontsize=16)
plt.xlabel("Days", fontsize=12)
plt.ylabel("Temperature", fontsize=12)
plt.legend(fontsize=10)
plt.grid(True, alpha=0.3)

plt.show()
````

#Scatter Plot - Height vs Weight! ðŸ“

````python
np.random.seed(42)
height = np.random.normal(170, 10, 50)
weight = height * 0.6 + np.random.randn(50) * 5

print("50 people data:")
print(f"Average height: {height.mean():.1f} cm")
print(f"Average weight: {weight.mean():.1f} kg")
````

````python
plt.scatter(height, weight, color='purple', alpha=0.6, s=100)

plt.title('Height vs Weight', fontsize=16)
plt.xlabel('Height (cm)', fontsize=12)
plt.ylabel('Weight (kg)', fontsize=12)
plt.grid(True, alpha=0.3)

plt.show()
````

````python
study_hours = np.array([2, 3, 1, 5, 4, 6, 2, 7, 3, 8,
                        4, 5, 1, 6, 3, 7, 4, 8, 5, 2,
                        6, 3, 7, 4, 8, 5, 6, 2, 7, 4])

scores = np.array([65, 70, 55, 80, 75, 85, 60, 90, 68, 95,
                   72, 78, 58, 83, 70, 88, 74, 93, 79, 64,
                   82, 69, 87, 76, 94, 81, 84, 62, 89, 73])
````

#Company Comparison

````python
months = np.arange(1, 7)
company_a = np.arange(100, 220, 20)
company_b = np.arange(90, 200, 20)
company_c = np.array([110, 115, 125, 135, 145, 160])

print("Company A:", company_a)
print("Company B:", company_b)
print("Company C:", company_c)
````

````python
plt.plot(months, company_a, label='Company A', color='red', linewidth=2)
plt.plot(months, company_b, label='Company B', color='blue', linewidth=2)
plt.plot(months, company_c, label='Company C', color='green', linewidth=2)

plt.title("Company Sales Comparison", fontsize=16)
plt.xlabel("Months", fontsize=12)
plt.ylabel("Sales ($1000)", fontsize=12)
plt.legend(fontsize=10)
plt.grid(True, alpha=0.3)

plt.show()
````

#Temperature Scatter

````python
temperature = np.random.uniform(20, 35, 40)
ice_cream_sales = temperature * 3 + np.random.randn(40) * 10

print("Temperature:", temperature)
print("Ice Cream Sales:", ice_cream_sales)
print(f"Average temperature: {temperature.mean():.2f}")
print(f"Average ice cream sales: {ice_cream_sales.mean():.2f}")
````

````python
plt.scatter(temperature, ice_cream_sales, color='red', alpha=0.6, s=100)

plt.title("Temperature vs Ice Cream Sales", fontsize=16)
plt.xlabel("Temperature (Â°C)", fontsize=12)
plt.ylabel("Ice Cream Sales (units)", fontsize=12)
plt.grid(True, alpha=0.3)

plt.show()
````

#Stock Prices

````python
days = np.arange(1, 11)
apple = np.array([150, 152, 148, 155, 153, 157, 160, 158, 162, 165])
google = np.array([2800, 2820, 2790, 2850, 2830, 2870, 2900, 2880, 2920, 2950])
````

````python
plt.plot(days, apple, color='gold', linewidth=2)

plt.title("Stock Prices", fontsize=16)
plt.xlabel("Days", fontsize=12)
plt.ylabel("Price Apple($)", fontsize=12)
plt.legend(fontsize=10)
plt.grid(True, alpha=0.3)

plt.show()
````

````python
plt.plot(days, google,  color='red', linewidth=2)

plt.title("Stock Prices", fontsize=16)
plt.xlabel("Days", fontsize=12)
plt.ylabel("Price Google ($)", fontsize=12)
plt.grid(True, alpha=0.3)

plt.show()
````

normalize qiling (percentage change)

````python
normlize_apple = (apple - apple.min()) / (apple.max() - apple.min())
normlize_google = (google - google.min()) / (google.max() - google.min())

print("Apple:", normlize_apple)
print("Google:", normlize_google)
````

````python
plt.plot(days, normlize_apple, label='Apple', color='gold', linewidth=2)
plt.plot(days, normlize_google, label='Google', color='red', linewidth=2)

plt.title("Stock Prices", fontsize=16)
plt.xlabel("Days", fontsize=12)
plt.ylabel("Percentage Change", fontsize=12)
plt.legend(fontsize=10)
plt.grid(True, alpha=0.3)

plt.show()
````
