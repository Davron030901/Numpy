RANDINT

```
np.random.randint(low, high, size)

 low: minimum (included)
 high: maximum (EXCLUDED!)
 size: number of choices

CHOICE

```
np.random.choice(array, size)

 array: list to choose from
 size: number of choices (optional)

#Dice Statistics

100 rocks

````python
dice = np.random.randint(1, 7, 100)
print("Dice:", dice)
````

Count how many times each number appears

````python
one = np.count_nonzero(dice == 1)
two = np.count_nonzero(dice == 2)
three = np.count_nonzero(dice == 3)
four = np.count_nonzero(dice == 4)
five = np.count_nonzero(dice == 5)
six = np.count_nonzero(dice == 6)
print(f"1: {one}, 2: {two}, 3: {three}, 4: {four}, 5: {five}, 6: {six}")
````

````python
one = (dice == 1).sum()
two = (dice == 2).sum()
three = (dice == 3).sum()
four = (dice == 4).sum()
five = (dice == 5).sum()
six = (dice == 6).sum()
print(f"1: {one}, 2: {two}, 3: {three}, 4: {four}, 5: {five}, 6: {six}")
````

#Restaurant Choice

````python
restaurants = ["Somsa House", 'Manti Palace', "Palav Centre", 'Lag\'mon Cafe', 'Shashlik Garden']
week = ["Monday", "Tuesday", "Wednesday", "Thursday", "Friday", "Saturday", "Sunday"]
print("Restaurants:", restaurants)
print("Week:", week)
````

1 restaurant per day

````python
for day in week:
  restaurant = np.random.choice(restaurants)
  print(f"{day}: {restaurant}")
````
