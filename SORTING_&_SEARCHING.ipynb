```
np.sort() - sort array (copy qaytaradi)
np.argsort() - sort indices
np.argmax() / np.argmin() - max/min indices
np.searchsorted() - binary search
Sorting 2D arrays - axis parameter

```
Key concepts:
SORT:

  Returns sorted copy
  Original unchanged
  Can sort ascending/descending

ARGSORT:

  Returns indices that would sort array
  Keep track of original positions
  Very useful for related arrays



````python
import numpy as np
````

#Basic Sorting! 🔍

````python
arr = np.array([3, 1, 4, 1, 5, 9, 2, 6, 5, 3])
print("Original array:")
print(arr)
````

SORT (ascending)

````python
sorted_arr = np.sort(arr)
print("Sorted array:")
print(sorted_arr)
````

SORT (descending)

````python
sorted_desc = np.sort(arr)[::1]
print("Sorted array (descending):")
print(sorted_desc)
````

Original unchanged

````python
print("Original (unchanged):")
print(arr)
````

#Argsort - Track Positions! 📍

````python
students = np.array(['Ali', 'Vali', 'Guli', 'Diyor', 'Aziza'])
scores = np.array([85, 92, 78, 95, 88])

print("Students:", students)
print("Scores:", scores)
````

Get indices that would sort scores

````python
sorted_indices = np.argsort(scores)
print("Indices that would sort scores:")
print(sorted_indices)
````

Sort by score (ascending)

````python
print("Students by score (low to high):")
for idx in sorted_indices:
  print(f"{students[idx]}: {scores[idx]}")
````

Sort by score (descending) - top performers

````python
sorted_indices_desc = np.argsort(scores)[::-1]
print("Top performers by score (high to low):")
for idx in sorted_indices_desc:
  print(f"{students[idx]}: {scores[idx]}")
````

#Argmax & Argmin! 🏆

````python
products = np.array(['Laptop', 'Phone', 'Tablet', 'Monitor', 'Keyboard'])
sales = np.array([45, 120, 80, 35, 95])

print("Products:", products)
print("Sales:", sales)
````

Find best seller (max)

````python
max_idx = np.argmax(sales)
print(f"Best seller: {products[max_idx]} ({sales[max_idx]} uints)" )
````

Find worst seller (min)

````python
min_idx = np.argmin(sales)
print(f"Worst seller: {products[min_idx]} ({sales[min_idx]} uints)" )
````

Find top 3 sellers

````python
top3_indices = np.argsort(sales)[3:][::-1]
print("Top 3 sales:")
for i, idx in enumerate(top3_indices, 1):
  print(f"{i}, {products[idx]}: ({sales[idx]} uints)")
````

#Sorting 2D Arrays! 📐

````python
scores = np.array([[85, 90, 78],
                   [92, 88, 95],
                   [78, 85, 82],
                   [95, 92, 90]])

print("Original scores:")
print(scores)
````

Sort each row (axis=1)

````python
sorted_scores = np.sort(scores, axis=1)
print("Sorted scores (each row):")
print(sorted_scores)
````

Sort each column (axis=0)

````python
sorted_cols = np.sort(scores, axis=0)
print("Scored (each column):")
print(sorted_cols)
````

Sort by total score

````python
total_scores = scores.sum(axis=1)
print("Total scores:")
print(total_scores)
````

````python
sorted_by_total = np.argsort(total_scores)[::-1]
print("Students ranked by total:")
for rank, idx in enumerate(sorted_by_total, 1):
  print(f"Rank {rank}: Student {idx+1} - Total: {total_scores[idx]}")
````

#Sales Ranking

````python
products = np.array(['A', 'B', 'C', 'D', 'E'])
sales = np.array([120, 85, 200, 95, 150])

print("Products:", products)
print("Sales:", sales)
````

````python
sorted = np.argsort(sales)[::-1]
print("Products id :", sorted)
````

````python
top3_product = sorted[:3]
print("Top 3 products:")
for idx in top3_product:
  print(f"{products[idx]} : {sales[idx]}")
````

````python
worst2 = np.argsort(sales)[::1]
print("Worst 2 products:")
for idx in worst2[:2]:
  print(f"{products[idx]} : {sales[idx]}")
````

````python
median_sales = np.median(sales)
print("Median sales:", median_sales)
````

#Grade Analysis

````python
grades = np.array([85, 92, 78, 95, 88, 76, 90, 82, 87, 93])

print("Original grades:", grades)
````

````python
sort = np.sort(grades)
print("Sorted grades:", sort)
````

````python
indices_top = np.argsort(grades)[::-1]
print("Indices of sorted grades:", indices_top)
````

````python
print("Top 3 students:")
for idx in indices_top[:3]:
  print(f"Student {idx+1}: {grades[idx]}")
````

````python
indices_bottom = np.argsort(grades)
print("Indices of sorted grades:", indices_bottom)
````

````python
print("Bottom 3 students:")
for idx in indices_bottom[:3]:
  print(f"Student {idx+1}: {grades[idx]}")
````

````python
middle_grade = np.median(grades)
print("Middle grade:", middle_grade)
````

#2D Performance Matrix

````python
performance = np.array([[85, 90, 88, 92],
                        [78, 82, 85, 88],
                        [92, 95, 93, 96]])
print("Original performance matrix:")
print(performance)
````

````python
sort_each_employee = np.sort(performance, axis=1)
print("Sorted performance (each employee):")
print(sort_each_employee)
````

````python
sort_row_ids = np.argsort(performance, axis=1)
print("Sorted row indices:")
print(sort_row_ids)
````

````python
sort_col_idx = np.argsort(performance, axis=0)
print("Sorted column indices:")
print(sort_col_idx)
````

````python
overal_best = np.argmax(performance)
print("Overall best performer:", overal_best)
````
