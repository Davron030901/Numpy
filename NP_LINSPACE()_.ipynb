#Why do you need linspace()?

1. Graphic drawing: 100 points from 0 to 2π (smooth sin graph)

```python
x=np.linspace(0,2*np.pi,100)
print(x)
```

2. Image size: Resize image from 256x256 to 512x512
 Requires exactly 512 coordinates

```python
coords=np.linspace(0,255,512)
print(coords)
```

3. Time series: 0 to 10 seconds, 50 FPS video

```python
time=np.linspace(0,10,500)
print(time)
```

# endpoint Parameter

Default: endpoint=True (takes the last number)

```python
arr1=np.linspace(0,10,5)
print("endpoint=True (default):")
print(arr1)
```

endpoint=False (does not get the last number)

```python
arr2=np.linspace(0,10,5,endpoint=False)
print("endpoint=False:")
print(arr2)
```

#Retstep - Know the step

When you need an interval in mathematical calculations!

```python
arr, step=np.linspace(0,10,5,retstep=True)

print("Array: ")
print(arr)

print(f"\n Qadam (step): {step}")
```
