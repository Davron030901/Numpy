#Real-world - Image Processing

```python
img = np.array([[100, 150, 200],
                [50, 100, 150],
                [0, 50, 100]], dtype=np.uint8)

print("Original Image:")
print(img)
```

Add BRIGHTNESS

```python
brighter= img + 50  # Overflow muammosi!
print("Brighter (unsafe):")
print(brighter)
```

SAFE brightness (clip)

```python
brighter_safe = np.clip(img + 50, 0, 255).astype(np.uint8)
print("Brighter (safe):")
print(brighter_safe)
```

CONTRAST (kontrast oshirish)

```python
contrast = np.clip(img * 1.5, 0, 255).astype(np.uint8)
print("Increased Contrast:")
print(contrast)
```

NEGATIVE (invert)

```python
negative = 255 - img
print("Negative Image:")
print(negative)
```

THRESHOLD (binary)

```python
treshold = (img > 100).astype(np.uint8) * 255
print("Thresholded Image:")
print(treshold)
```
