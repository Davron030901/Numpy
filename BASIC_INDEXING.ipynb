#3D Indexing - RGB Image

##4x4 RGB image (small example)

Shape: (4, 4, 3) = 4 height, 4 width, 3 channels

img [row, column, channel]

                   

      bo'y, en,    kanal (0=R, 1=G, 2=B)



```python
img=np.array([
    [[255, 0, 0], [0, 255, 0], [0, 0, 255], [255, 255, 0]],  #row 0
    [[255, 0, 255], [0, 255, 255], [128, 128, 128], [0, 0, 0]],  #row 1
    [[255, 128, 0], [128, 0, 255], [0, 128, 255], [255, 255, 255]],  #row2
    [[100, 100, 100], [200, 200, 200], [50, 150, 250], [10, 20, 30]] #row 3
])
print(img)
```

```python
print("RGB Image shape:", img.shape)
```

ONE PIXEL, ALL CHANNELS

```python
print("Pixel [0, 0] (red):")
print(img[0, 0]) # [255, 0, 0] - R=255,  G=0,  B=0
```

ONE PIXEL, ONE CHANNEL

```python
print("Pixel [0, 0],  Red channel:")
print(img[0, 0, 0]) # 255 - faqat R

print("Pixel [0, 0],  Green channel:")
print(img[0, 0, 1]) # 0 - faqat G

print("Pixel [0,  0],  Blue channel:")
print(img[0, 0, 2]) # 0 - faqat B
```

CENTER PIXEL

```python
center_y=img.shape[0]//2
center_x=img.shape[1]//2
print(f"Center pixel [{center_y},{center_x}]:")
print(img[center_y, center_x])
```

10x10 rasm yarating, diagonaldan element oling

```python
arr = np.arange(1,101).reshape(10,10)
print("10x10 Array:")
print(arr)
```

For loop

```python
for i in range(10):
  print(f"arr[{i}, {i}] = {arr[i,i]}")
```

List comprehension (advanced!)

```python
diagonal_elements =[arr[i, i] for i in range(10)]
print("Diagonal elements:", diagonal_elements)
```

NumPy diagonal function

```python
dioganal_np = np.diag(arr)
print("Diagonal:",dioganal_np)
```
