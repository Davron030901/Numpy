#3 months, 4 weeks of trading

```python
monthly = np.array([[20,    25,     22,     28],
                    [22,    27,     24,     30],
                    [21,    26,     23,     29]])
weeks = ["Week 1", "Week 2", "Week 3", "Week 4"]
print("Monthly sales:")
print(monthly)
```

```python
for week, avg in zip(weeks, monthly.mean(axis=0)):
  print(f"{week} average: {avg}")
```

#4 students, 3 subject grades

```python
grades = np.array([[85,   78,     92],
                   [90,   82,     88],
                   [78,   85,     95],
                   [88,   80,     90]])
students = ["Ali", "Vali", "Guli", "Diyor"]
print("Grades:")
print(grades)
```

```python
for student, avg in zip(students, grades.mean(axis=1)):
  print(f"{student} avg:{avg:.2f}")
```

#5 stores, 4 products (daily sales)

```python
sales = np.array([[50,   40,    30,    35],
                  [45,   38,    28,    32],
                  [52,   42,    32,    38],
                  [48,   39,    29,    34],
                  [51,   41,    31,    36]])
products = ["Apple", "Banana", "Cherry", "Durian"]
stores = ["Store A", "Store B", "Store C", "Store D", "Store E"]
print("Sales:")
print(sales)
```

Total sales of each PRODUCT

```python
total_products = sales.sum(axis=0)
for product, total in zip(products, total_products):
  print(f"{product} total: {total}")
```

```python
total_stores = sales.sum(axis=1)
for store, total in zip(stores, total_stores):
  print(f"{store} total: {total}")
```

```python
product_max_idx = total_products.argmax()
print(f"{products[product_max_idx]}: ${total_products[product_max_idx]}")
```

```python
store_max_idx = total_stores.argmax()
print(f"{stores[store_max_idx]}: ${total_stores[store_max_idx]}")
```
