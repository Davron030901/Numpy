````python
import numpy as np
````

# Creating Dates! 📅

````python
dates = np.array(['2024-01-15', '2024-06-20', '2024-12-25'],
                 dtype='datetime64')
print("Dates from strings:")
print(dates)
print(f"Type: {dates.dtype}")
````

Today

````python
today = np.datetime64('today')
print(f"Today: {today}")
````

Date range

````python
jan_dates = np.arange('2024-01-01', '2024-01-08', dtype='datetime64[D]')
print("January dates:")
print(jan_dates)
````

Monthly range

````python
months = np.arange('2024-01', '2024-07', dtype='datetime64[M]')
print("Firs 6 months of 2024:")
print(months)
````

#Date Arithmetic! ➕➖

````python
start = np.datetime64('2024-01-01')
print(f"Start date: {start}")
````

ADD days

````python
one_day = np.timedelta64(1, 'D')
tomorrow = start + one_day
print(f"One day: {tomorrow}")
````

````python
one_week = np.timedelta64(7, 'D')
next_week = start + one_week
print(f"Next week: {next_week}")
````

````python
start_date = start.astype('datetime64[M]')
one_month = np.timedelta64(1, 'M')
next_month = start_date + one_month
print(f"Next month: {next_month}")
````

SUBTRACT (time difference)

````python
end = np.datetime64('2024-12-31')
duration = end - start
print(f"Duration: {duration}")
````

````python
dates = np.array(['2024-01-01', '2024-02-01', '2024-03-01'],
                 dtype='datetime64')

later = dates + np.timedelta64(15, 'D')
print("Original dates:")
print(dates)
print("Later dates:")
print(later)
````

#Date Comparison! ⚖️

````python
events = np.array(['2024-01-15', '2024-03-20', '2024-06-10',
                   '2024-09-05', '2024-12-25'], dtype='datetime64')
print("Event dates:")
print(events)
````

Current date

````python
today = np.datetime64('2024-06-01')
print(f"Today: {today}")
````

PAST events (before today)

````python
past_mask = events < today
past_events = events[past_mask]
print("Past events:")
print(past_events)
print(f"Count: {past_mask.sum()}")
````

FUTURE events (after today)

````python
future_mask = events > today
future_events = events[future_mask]
print("Future events:")
print(future_events)
print(f"Count: {future_mask.sum()}")
````

Days until future events

````python
days_until = future_events - today
print("Days until future events:")
for event, days in zip(future_events, days_until):
  print(f"{event}: {days.astype('timedelta64[D]')}")
````

#Real Example - Project Timeline! 📊

````python
project_start = np.datetime64('2024-01-01')
print(f"Project start: {project_start}")
````

````python
milestones = np.array([30, 60, 90, 120, 150])  # Days
milestone_names = ['Planning', 'Development', 'Testing', 'Review', 'Launch']
````

````python
milestone_dates = project_start + milestones * np.timedelta64(1, 'D')

print("Project milestones:")
for name, date, days in zip(milestone_names, milestone_dates, milestones):
  print(f'{name:12} - Day {days:3} - {date}')
````

Today's status

````python
today = np.datetime64('today')
print(f"Today: {today}")
````

````python
completed_mask = milestone_dates <= today
completed = milestone_dates[completed_mask]
print(f'Completed milestones:{completed_mask.sum()}/{len(milestones)}')
for i, name in enumerate(milestone_names):
  if completed_mask[i]:
    print(f'{name}')

  else:
    days_remaining = (milestone_dates[i] - today).astype('timedelta64[D]')
    print(f"{name} - {days_remaining} remaining")
````

#Birthday Analysis

````python
birthdays = np.array(['2000-03-15', '1995-07-20', '2002-11-05',
            '1998-01-10', '2001-09-25'], dtype='datetime64')
print("Birthdays:")
print(birthdays)
````

````python
today = np.datetime64('2024-02-22')
print(f"Today: {today}")
````

````python
years = (today - birthdays) / 365
print("Years until birthdays:")
print(years)
````

````python
large_25 = years > 25
print("Large birthdays:")
print(large_25)
print(f"Count: {large_25.sum()}")
````

````python
oldest = years.max()
print(f"The oldest: {oldest}")
youngest = years.min()
print(f"Youngest: {youngest}")
````

#Deadline Tracker

````python
tasks = np.array(['Report', 'Presentation', 'Code Review', 'Testing'])
deadlines = np.array(['2024-02-25', '2024-03-05', '2024-02-20', '2024-03-10'],dtype='datetime64')
print("Tasks and deadlines:")
print(tasks)
print(deadlines)
today = np.datetime64('2024-02-22')
print(f"Today: {today}")
````

````python
overdue_task = deadlines < today
print("Overdue tasks:")
print(overdue_task)
print(tasks[overdue_task])
````

````python
upcoming_task = deadlines >= today
print("Upcoming tasks:")
print(upcoming_task)
print(tasks[upcoming_task])
````

````python
print("Status table (Task| Deadline | Status):")
for task, deadline in zip(tasks, deadlines):
  status = 'Overdue' if deadline < today else 'Upcoming'
  print(f"task: {task}, deadline: {deadline}, status: {status}")
````
