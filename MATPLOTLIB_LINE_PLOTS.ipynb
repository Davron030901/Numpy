```
function:

  plt.plot() - draw a line
  plt.title() - title
  plt.xlabel(), plt.ylabel() - axis labels
  plt.grid() - grid
  plt.show() - show

````python
import matplotlib.pyplot as plt
import numpy as np
````

````python
x = [1, 2, 3, 4, 5]
y = [2, 4, 6, 8, 10]
````

````python
plt.plot(x, y)
plt.show()
````

**Basic Workflow:**
```
1. Import → import matplotlib.pyplot as plt
2. Data   → x, y arrays
3. Plot   → plt.plot(x, y)
4. Style  → plt.title(), plt.xlabel(), etc.
5. Show   → plt.show()

#Simple Line

````python
x = np.array([1, 2, 3, 4, 5])
y = np.array([2, 4, 6, 8, 10])

print("X:", x)
print("Y:", y)
````

````python
plt.plot(x, y)

plt.title("My First Plot")

plt.xlabel("X values")
plt.ylabel("Y values")

plt.grid(True)

plt.show()
````

#Sin Wave - Beautiful! 🌊

````python
x = np.linspace(0, 10, 100)
y= np.sin(x)

print("X 100 points range [0, 10]")
print("Y sin(x) values")
````

````python
plt.plot(x, y)

plt.title("Sin Wave", fontsize=16)
plt.xlabel("X", fontsize=12)
plt.ylabel("sin(x)", fontsize=12)
plt.grid(True, alpha=0.3)

plt.show()
````

#Weekly Temperature - Real Data! 🌡️

````python
days = np.array([1, 2, 3, 4, 5, 6, 7])
temperature = np.array([20, 22, 25, 23, 21, 24, 26])

day_names = ["Mon", "Tue", "Wed", "Thu", "Fri", "Sat", "Sun"]

print("Days:", day_names)
print("Temperature:", temperature)
````

````python
plt.plot(day_names, temperature, color="red", linewidth=2)

plt.title("Weekly Temperature", fontsize=16)
plt.xlabel("Day", fontsize=12)
plt.ylabel("Temperature (°C)", fontsize=12)
plt.grid(True, linestyle="--", alpha=0.5)

plt.show()
````

#Multiple Customizations! 🎨

````python
x = np.linspace(0, 5, 50)
y = x ** 2
````

````python
plt.plot(x, y,
         color="green",
         linewidth=3,
         linestyle="-",
         marker="o",
         markersize=4,
         alpha=0.7)

plt.title("Custom Plot", fontsize=18)
plt.xlabel("X", fontsize=14)
plt.ylabel("Y", fontsize=14)
plt.grid(True, color="gray",linestyle=":", alpha=0.5)

plt.show()
````

#Linear Growth

````python
years = np.arange(2019, 2025)
benifit = np.array([100, 150, 200, 280, 350, 450])

print("Years:", years)
print("Benifit:", benifit)
````

````python
plt.plot(years, benifit, color="blue", linewidth=2)

plt.title("Company Growth", fontsize=16)
plt.xlabel("Years", fontsize=12)
plt.ylabel("Benifit (thousand $)", fontsize=12)
plt.grid(True, alpha=0.3)

plt.show()
````

#Cos Wave

````python
x = np.linspace(0, 2*np.pi, 100)
y = np.cos(x)
````

````python
plt.plot(x, y, color="purple", linewidth=2)

plt.title("Cos Wave", fontsize=16)
plt.xlabel("X", fontsize=12)
plt.ylabel("cos(x)", fontsize=12)
plt.grid(True, alpha=0.3)

plt.show()
````

#Student Progress

````python
tests = np.arange(1, 8)
grades = np.array([60, 65, 70, 75, 80, 85, 90])
````

````python
plt.plot(tests, grades, color='orange', linewidth=2, marker="o")

plt.title("Student Progress", fontsize=16)
plt.xlabel("Test Number", fontsize=12)
plt.ylabel("Grade", fontsize=12)
plt.grid(True, linestyle="--", alpha=0.5)

plt.show()
````
