#Normal Distribution! 🔔

Standard normal (mean=0, std=1)

````python
print("1. np.random.randn() - standard normal")
standard = np.random.randn(10)
print(standard)
print(f"Min: {standard.min():.3f}, Max: {standard.max():.3f}")
````

Custom normal distribution

````python
print("2. np.random.normal() - custom parameters")

scores = np.random.normal(75, 10, size=20)
print("Test scores:")
print(scores)
print(f"Mean: {scores.mean():.2f}, Std: {scores.std():.2f}")
````

Grade distribution

````python
print("3. Grade distribution:")
grade_ranges = [
    (scores < 60, 'F'),
    ((scores >= 60) & (scores < 70), 'D'),
    ((scores >= 70) & (scores < 80), 'C'),
    ((scores >= 80) & (scores < 90), 'B'),
    (scores >= 90, 'A')
]
for condition, grade in grade_ranges:
  count = condition.sum()
  print(f"Grade {grade}: {count} students ({count/len(scores)*100:.0f}%)")
````

#Random Integers & Choice! 🎯

Random integers

````python
print("1. np.random.randint()")
dice = np.random.randint(1, 7, size=20)
print("Dice rolls:")
print(dice)
print(f"Unique values: {np.unique(dice)}")
print(f"Min: {dice.min()}, Max: {dice.max()}")
print(f"Average: {dice.mean():.2f}")
````

Count each number

````python
print("Dice frequency:")
for num in range(1, 7):
  count = (dice == num).sum()
  print(f"Number {num}:{'█' * count} {count} times")
````

Random choice from array

````python
print("2. np.random.choice()")
students = ['Alice', 'Bob', 'Charlie', 'David', 'Eve', 'Frank']
team = np.random.choice(students, size=3, replace=False)
print("Students:")
print(students)
print("Team:")
print(team)
````

Weighted choice

````python
print("3. Weighted random choice")
options = ['A', 'B', 'C']
probabilities = [0.5, 0.3, 0.2]
result = np.random.choice(options, size=100, p=probabilities)
print("Results (100 trials):")
for option in options:
  count = (result == option).sum()
  print(f"{option}: {count} times ({count}%)")
````

# Reproducibility with Seed! 🌱

WITHOUT seed - different every time

````python
print("1. WITHOUT seed (random):")
print("Run 1:", np.random.rand(5))
print("Run 2:", np.random.rand(5))
````

 WITH seed - same every time

````python
print("2. WITH seed (reproducible):")
np.random.seed(42)
print("Run 1:", np.random.rand(5))
np.random.seed(42)
print("Run 2:", np.random.rand(5))
print("(Identical!)")
````

Practical use: testing

````python
print("3. Practical example - Unit testing")
def simulate_experiment(seed=None):
  if seed is not None:
    np.random.seed(seed)

    flips = np.random.choice(['Heads', 'Tails'], size=100)
    heads = (flips == 'Heads').sum()
    return heads
````

Same seed = same result

````python
result1 = simulate_experiment(seed=123)
result2 = simulate_experiment(seed=123)

print("Result 1:", result1)
print("Result 2:", result2)
print(f"Result match: {result1 == result2}")
````

#Monte Carlo Simulation

````python
np.random.seed(200)
x, y = np.random.uniform(0, 1, size=10000), np.random.uniform(0, 1, size=10000)
print("Points x:")
print(x)
print("Points x:")
print(y)
````

````python
distanse = np.sqrt(x**2 + y**2)
print("Distance:")
print(distanse)
````

````python
inside_circle = distanse < 1
print("Points inside circle:")
print(inside_circle)
````

````python
inside = inside_circle.sum()
print(f"Number of points inside the unit circle: {inside}")
````

````python
total = len(distanse)
print("Len total:", total)
````

````python
pi = 4 * inside / total
print("Estimated value of pi:", pi)
````

````python
compare = np.pi
````

````python
print(f"Estimated Pi: {pi}")
print(f"Actual Pi (np.pi): {compare}")
print(f"Difference: {abs(pi - compare):.4f}")
````
