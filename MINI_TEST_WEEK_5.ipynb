#Fancy Indexing + Sorting

```python
employee_ids = np.array([101, 102, 103, 104, 105, 106, 107, 108])
performance = np.array([78, 92, 85, 95, 88, 76, 90, 82])

print("IDs:", employee_ids)
print("Performance:", performance)
```

```python
top_per = np.argsort(performance)[::-1]
print("Top 3 performance:", top_per[:3])
```

```python
high_per_ids = np.where(performance >= 90)
print(high_per_ids[0])
```

```python
print("Top 3 employees performanse: ")
for idx in top_per[:3]:
  print(f"IDs:{employee_ids[idx]}  , performance: {performance[idx]}")
```

```python
count_high_per = high_per_ids[0].size
print("High performers (>=90) count:", count_high_per)
```

#Boolean + Ufuncs

```python
data = np.array([25, 100, 64, 144, 81, 49, 121, 36])

print("Data:", data)
```

```python
root = np.sqrt(data)
print("Square root:", root)
```

Boolean mask

```python
boolen_mask = root < 10
print("Boolean mask:", boolen_mask)
```

```python
small_values = data[boolen_mask]
print("Small values:", small_values)
```

```python
mean = small_values.mean()
print("Mean:", mean)
```

#Broadcasting + Normalization

```python
scores = np.array([[85, 90, 78, 92],
                  [88, 85, 82, 95],
                  [79, 88, 90, 87]])

print("Scores:")
print(scores)
```

```python
mean = scores.mean(axis=0)
print("Mean:", mean)
```

```python
std = scores.std(axis=0)
print("Standard deviation:", std)
```

```python
z_scores = (scores - mean) / std
print("Z-scores:")
print(z_scores)
```

```python
verify_mean = z_scores.mean(axis=0)
print("Verify mean:", verify_mean)
```

```python
verify_std = z_scores.std(axis=0)
print("Verify standard deviation:", verify_std)
```

#2D Sorting + Argmax

```python
sales = np.array([[120, 150, 130],
                  [200, 180, 190],
                  [80, 95, 85],
                  [150, 140, 160]])

print("Sales:")
print(sales)
```

```python
sort_month = np.sort(sales, axis=1)
print("Sort month:")
print(sort_month)
```

```python
best_month_each_product = np.argmax(sales, axis=1)
print("Best month product:")
print(best_month_each_product)
```

```python
product_each_month = np.argmax(sales, axis=0)
print("Product each month:")
print(product_each_month)
```

```python
overal_best_month = np.argmax(sales)
print("Overal best month:", overal_best_month)
```

```python
values = sales.max()
print("Values:", values)
```

#Combined Challenge

```python
temperature = np.array([[22, 18, 25],
                        [25, 20, 28],
                        [28, 22, 30],
                        [24, 19, 26],
                        [26, 21, 29],
                        [23, 17, 27],
                        [27, 23, 31]])

print("Temperature:")
print(temperature)
```

```python
avg_temperature_city = temperature.mean(axis=0)
print("Average temperature city:")
print(avg_temperature_city)
```

```python
comfortable_range = (avg_temperature_city >= 20) & (avg_temperature_city <= 25)
print("Boolean: comfortable range 20-27°C")
print(comfortable_range)
```

```python
comfortable_range = (temperature >= 20) & (temperature <= 25)
print("Boolean: comfortable range 20-27°C")
print(comfortable_range)
```

```python
valid_range_clip = np.clip(temperature, 15, 35)
print("Valid range clip:")
print(valid_range_clip)
```

```python
col_to_hot = np.argsort(avg_temperature_city)
print("Column to hot:")
print(col_to_hot)
```
