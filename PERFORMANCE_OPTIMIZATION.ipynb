````python
import numpy as np
import time
````

````python
# SLOW: Python loop
start = time.time()
result = []
for i in range(1000000):
    result.append(i * 2)
loop_time = time.time() - start

# FAST: Vectorized
start = time.time()
result = np.arange(1000000) * 2
vector_time = time.time() - start

print(f"Loop: {loop_time:.4f}s")
print(f"Vectorized: {vector_time:.4f}s")
print(f"Speedup: {loop_time / vector_time:.1f}x faster!")
````

#Pre-allocation! 📦

````python
n = 100_000
````

Growing array

````python
start = time.time()
arr_bad = np.array([])
for i in range(n):
    arr_bad = np.append(arr_bad, i)
bad_time = time.time() - start

print(f"BAD - Growing array ({n:,} elements):")
print(f"Time: {bad_time:.4f}s")
````

Pre-allocated

````python
start = time.time()
arr_good = np.zeros(n)
for i in range(n):
  arr_good[i] = i
good_time = time.time() - start

print(f"GOOD - Pre-allocated:")
print(f"Time: {good_time:.4f}s")
print(f"Speedup: {bad_time / good_time:.1f}x faster!")
````

No loop at all!

````python
start = time.time()
arr_best = np.arange(n)
best_time = time.time() - start

print(f"BEST - Vectorized (no loop):")
print(f"Time: {best_time:.4f}s")
print(f"vs pre-allocated: {good_time / best_time:.1f}x faster!")
print(f"vs growing: {bad_time / best_time:.0f}x faster!")
````

#Broadcasting Efficiency! 📡

````python
matrix = np.random.rand(1000, 1000)
vector = np.random.rand(1000)

print(f"Matrix shape: {matrix.shape}")
print(f"Vector shape: {vector.shape}")
````

METHOD 1: Loop (slow)

````python
start = time.time()
result_loop = np.zeros_like(matrix)
for i in range(matrix.shape[0]):
  result_loop[i] = matrix[i] + vector
loop_time = time.time() - start

print(f"Loop method:")
print(f"Time: {loop_time:.4f}s")
````

METHOD 2: Broadcasting (fast!)

````python
strat = time.time()
result_broadcast = matrix + vector
broadcast_time = time.time() - start

print(f"Broadcasting:")
print(f"Time: {broadcast_time:.4f}s")
print(f"Speedup: {loop_time / broadcast_time:.1f}x faster!")
````

````python
print(f"Results equal: {np.allclose(result_loop, result_broadcast)}")
````

#Contiguous Memory! 🧱

````python
arr = np.random.rand(1000, 1000)
print(f"Original C-contiguous: {arr.flags['C_CONTIGUOUS']}")

arr_t = arr.T
print(f"Transpose C-contiguous: {arr_t.flags['C_CONTIGUOUS']}")
print(f"Transpose F-contiguous: {arr_t.flags['F_CONTIGUOUS']}")
````

Operation on non-contiguous

````python
start = time.time()
result_t = arr_t.sum()
time_t = time.time() - start

arr_copy = np.ascontiguousarray(arr_t)
print(f"Copy C-contiguous: {arr_copy.flags['C_CONTIGUOUS']}")
````

Operation on contiguous

````python
start = time.time()
result_copy = arr_copy.sum()
time_copy = time.time() - start

print(f"Non-contiguous sum: {time_t:.6f}s")
print(f"Contiguous sum: {time_copy:.6f}s")
print(f"Speedup: {time_t / time_copy:.2f}x")
````

Matrix Operations Speed

````python
A = np.random.rand(500,500)
B = np.random.rand(500,500)
````

````python
mult_arr = []
start = time.time()
for i in range(len(A)):
  for j in range(len(B)):
    mult_arr.append(A[i] * B[j])
mult_time = time.time() - start

print(f"Result time: {mult_time:.2f}s")
````

````python
start = time.time()
broadcast_arr = A @ B
brodcast_time = time.time() - start

print(f"Result time: {brodcast_time:.2f}s")
````

````python
print(f"Speedup {mult_time / broadcast_time}x faster")
````
