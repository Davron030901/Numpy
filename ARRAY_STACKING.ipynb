#Basic Stacking!

````python
a = np.array([1, 2, 3])
b = np.array([4, 5, 6])

result = np.vstack((a, b))
print("VSTACK")
print(result)

result = np.hstack((a, b))
print("HSTACK")
print(result)
````

````python
c = np.array([7, 8, 9])

print("Array a:", a)
print("Array b:", b)
print("Array c:", c)
````

VERTICAL STACK

````python
vertical = np.vstack([a, b, c])
print("Vertical Stack:")
print(vertical)
print(f"Shape: {vertical.shape}")
````

HORIZONTAL STACK

````python
horizontal = np.hstack([a, b, c])
print("Horizontal Stack:")
print(horizontal)
print(f"Shape: {horizontal.shape}")
````

#Student Data - Real Example! ðŸ“š

3 student grades (Math, Physics, English)

````python
student1 = np.array([85, 90, 78])
student2 = np.array([88, 92, 94])
student3 = np.array([90, 88, 86])

print("Student 1:", student1)
print("Student 2:", student2)
print("Student 3:", student3)
````

VSTACK - combine all students

````python
class_data = np.vstack([student1, student2, student3])
print("Class Data:")
print(class_data)
print(f"Shape: {class_data.shape}")
````

Now we can do operations

````python
print("Subject averages (axis=0):")
subject_avg = class_data.mean(axis=0)
subjects = ['Math', 'Physics', 'English']

for subject, avg in zip(subjects, subject_avg):
  print(f"{subject}: {avg:.1f}")
````

````python
print("Student averages (axis=1):")
student_avg = class_data.mean(axis=1)
students = ['Student 1', 'Student 2', 'Student 3']

for student, avg in zip(students, student_avg):
  print(f"{student}: {avg:.1f}")
````

#Add New Column (hstack)! ðŸ“Š

Student scores (3 students, 3 subjects)

````python
scores = np.array([[85, 90, 78],
                   [88, 92, 84],
                   [90, 88, 86]])

print("Original scores:")
print(scores)
print(f"Shape: {scores.shape}")
````

````python
averages = scores.mean(axis=1).reshape(-1, 1)
print("Averages (reshaped to column):")
print(averages)
print(f"Shape: {averages.shape}")
````

HSTACK - add average column

````python
scores_with_avg = np.hstack([scores, averages])

print("Scores with Averages:")
print(scores_with_avg)
print(f"Shape: {scores_with_avg.shape}")

cols = ['Math', 'Physics', 'English', 'Average']
print("Columns:", cols)
````

#Daily Data Accumulation! ðŸ“…

````python
day1 = np.array([100, 150, 120, 180])
print("Day 1 sales:", day1)

weekly_sales = day1.reshape(1, -1)
print("Weekly sales (reshaped to row):")
print(weekly_sales)
print(f"Shape: {weekly_sales.shape}")
````

````python
day2 = np.array([110, 145, 125, 175])
print("Day 2 sales:", day2)

weekly_sales = np.vstack([weekly_sales, day2])
print("Weekly sales (after day 2):")
print(weekly_sales)
````

````python
day3 = np.array([105, 155, 130, 185])
print("Day 3 sales:", day3)

weekly_sales = np.vstack([weekly_sales, day3])
print("Weekly sales (after day 3):")
print(weekly_sales)
````

````python
print("Total per stone (axis=0):")
print(weekly_sales.sum(axis=0))
````

````python
print("Daily totals (axis=1):")
print(weekly_sales.sum(axis=1))
````

#Temperature Data

Weekly temperature of 3 cities (7 days)

````python
toshkent = np.array([20, 22, 25, 23, 21, 24, 26])
samarqand = np.array([18, 20, 22, 21, 19, 22, 24])
buxoro = np.array([22, 24, 27, 25, 23, 26, 28])
````

````python
cities_temperature = np.vstack([toshkent, samarqand, buxoro])
print("Cities Temperature:")
print(cities_temperature)
print(f'Shape: {cities_temperature.shape}')
````

````python
avg_daily = cities_temperature.mean(axis=0)
print("Average temperature (axis=0):")
print(avg_daily)
````

````python
avg_cities = cities_temperature.mean(axis=1)
print("Average temperature (axis=1):")
print(avg_cities)
````

#Add ID Column

Student grade (4 students, 3 subjects)

````python
scores = np.array([[85, 90, 78],
          [88, 92, 84],
          [90, 88, 86],
          [82, 85, 80]])

id = np.array([1, 2, 3, 4]).reshape(-1, 1)

print("Original scores:")
print(scores)
print(f"Shape: {scores.shape}")
print("ID column:")
print(id)
````

````python
students_scores = np.hstack((id, scores))
print("Students Scores:")
print(students_scores)
````

#Combine Teams

2 football team scores (5 games)

````python
team_a = np.array([2, 3, 1, 2, 3])
team_b = np.array([1, 2, 2, 3, 1])

print("Team A scores:", team_a)
print("Team B scores:", team_b)
````

````python
scores_team = np.vstack((team_a, team_b))
print("Scores Team:")
print(scores_team)
````

````python
all_goals = scores_team.sum(axis=0)
print("All goals (axis=0):")
print(all_goals)
````

````python
all_team_goals = scores_team.sum(axis=1)
print("All team goals (axis=1):")
print(all_team_goals)
````

````python
team = np.array(['Team A', 'Team B'])
print("Teams:", team)
max_goals = all_team_goals.max()
print("Max goals:", max_goals)
axis_max_goals = np.argmax(all_team_goals)
print("Axis max goals:", axis_max_goals)
print("Winner:", team[axis_max_goals])
````
