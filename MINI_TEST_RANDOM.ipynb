#Basic rand()

Generate 8 random floats

```python
numbers = np.random.rand(8)
print("Generate 8 random floats:")
print(numbers)
```

find and print the maximum value

```python
max_number = numbers.max()
print("Max number:")
print(max_number)
```

#Seed Reproducibility

```python
np.random.seed(100)
numbers1 = np.random.rand(4)
print("Generate 4 random floats:")
print(numbers1)
```

```python
np.random.seed(100)
numbers2 = np.random.rand(4)
print("Generate 4 random floats:")
print(numbers2)
```

Verify they are the same


```python
print("Same:", numbers1 == numbers2)
```

#randint() Application

Simulate rolling 2 dice 20 times

```python
dice = np.random.randint(1, 7, size=(20, 2))
print("Simulate rolling 2 dice 20 times:")
print(dice)
```

Calculate the sum for each pair

```python
sum_dice = dice.sum(axis=1)
print("Calculate the sum for each pair:")
print(sum_dice)
```

Print how many times you got 7 (lucky number!)

```python
lucky_number = (sum_dice==7).sum()
print("Print how many times you got 7 (lucky number!):")
print(lucky_number)
```

#choice() Unique

Shuffle all 6 people (unique selection)

```python
people = ['Ali', 'Vali', 'Guli', 'Diyor', 'Zebo', 'Kamol']
shuffle_people = np.random.choice(people, 6, replace=False)
print("Shuffle all 6 people (unique selection):")
print(shuffle_people)
```

Create pairs: person[0] gives to person[1],
   person[1] gives to person[2], ... person[5] gives to person[0]

```python
for i in range(len(shuffle_people)):
  print(f"{shuffle_people[i]} gives to {shuffle_people[(i+1)%len(shuffle_people)]}")
```
