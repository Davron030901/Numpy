````python
import numpy as np
````

#Mathematical Ufuncs! üî¢

````python
arr = np.array([1, 4, 9, 16, 25])
print("Original array:")
print(arr)
````

SQUARE ROOT

````python
sqrt = np.sqrt(arr)
print("Square root of the array(sqrt):")
print(sqrt)
````

SQUARE

````python
square = np.square(arr)
print("Square of the array(square):")
print(square)
````

EXPONENTIAL

````python
small_arr = np.array([1, 2, 3])
exp = np.exp(small_arr)
print("Exponential of the array(exp):")
print(exp)
````

LOGARITHM (natural log)

````python
log_arr = np.array([1, 10, 100, 1000])
log = np.log(log_arr)
print("Natural log of the array(log):")
print(log)
````

LOG10

````python
log10 = np.log10(log_arr)
print("Log base 10(log10):")
print(log10)
````

POWER

````python
base = np.array([2, 3, 4])
power = np.power(base, 3)
print("Power (base^3):")
print(power)
````

#Trigonometric Ufuncs! üìê

````python
angles_deg = np.array([0, 30, 45, 60, 90])
print("Angles (degrees):")
print(angles_deg)
````

Convert to radian

````python
angles_rad = np.deg2rad(angles_deg)
print("Angles (radians):")
print(angles_rad)
````

SINE

````python
sines = np.sin(angles_rad)
print("Sine values:")
for deg, sine in zip(angles_deg, sines):
  print(f"sin({deg}¬∞) = {sine:.4f}")
````

COSINE

````python
cosines = np.cos(angles_rad)
print("Cosine values:")
for deg, cosine in zip(angles_deg, cosines):
  print(f"cos({deg}¬∞) = {cosine:.4f}")
````

TANGENT

````python
tangents = np.tan(angles_rad)
print("Tangent values:")
for deg, tangent in zip(angles_deg, tangents):
  print(f"tan({deg}¬∞) = {tangent}")
````

#Comparison & Rounding! üîÑ

````python
a = np.array([1, 5, 3, 8, 2])
b = np.array([2, 4, 6, 7, 9])

print("Array a:", a)
print("Array b:", b)
````

MAXIMUM (element-wise max)

````python
maximum = np.maximum(a, b)
print("Maximum (element-wise):")
print(maximum)
````

MINIMUM (element-wise min)

````python
minimum = np.minimum(a, b)
print("Minimum (element-wise):")
print(minimum)
````

Values with decimals

````python
values = np.array([1.2, 2.5, 3.7, 4.1, -2.3, -1.8])
print("Original values:")
print(values)
````

ROUND (to nearest)

````python
rounted = np.round(values)
print("Rounded values:")
print(rounted)
````

FLOOR (round down)

````python
floor = np.floor(values)
print("Floor (round down):")
print(floor)
````

CEIL (round up)

````python
ceil = np.ceil(values)
print("Ceil (round up):")
print(ceil)
````

TRUNC (remove decimals)

````python
trunc = np.trunc(values)
print("Trunc (remove decimals):")
print(trunc)
````

Absolute & Clip! üìä

````python
arr = np.array([-5, -2, 0, 3, 7, -10, 15])
print("Original array:")
print(arr)
````

ABSOLUTE VALUE

````python
abs_arr = np.abs(arr)
print("Absolute values:")
print(abs_arr)
````

SIGN (-1, 0, or 1)

````python
signs = np.sign(arr)
print("Sign of each element:")
print(signs)
````

CLIP (limit to range)

````python
print("Clip to range [0, 10]:")

clipped = np.clip(arr, 0, 10)
print(clipped)
````

temperature clipping

````python
temperatures = np.array([18, 25, 32, 15, 28, 35, 10])
print("Temperatures:", temperatures)
print("Clip to comfortable range [20, 30]:")
comfortable = np.clip(temperatures, 20, 30)
print(comfortable)
````

#Data Normalization

````python
data = np.array([100, 150, 200, 250, 300])
print("Data:")
print(data)
````

Square root normalization

````python
sqrt_normalization = np.sqrt(data)
print("Sqrt normalization:")
print(sqrt_normalization)
````

Log normalization

````python
log_normalization = np.log10(data)
print("Log normalization:")
print(log_normalization)
````

Min-max normalization

````python
min = data.min()
max = data.max()
min_max_normalization = (data - min) / (max - min)
print("Min-max normalization to [0, 1]:")
print(min_max_normalization)
````

#Circle Calculations

````python
radii = np.array([1, 2, 3, 4, 5])
print("Circle radii:")
print(radii)
````

Calculate area

````python
area = np.pi * np.square(radii)
print("Area:")
for r, s in zip(radii, area):
  print(f"raius: {r}, area: {s}")
````

Calculate circumference

````python
circumference = 2 * np.pi * radii
print(circumference)
print("Circumference:")
for r, c in zip(radii, circumference):
  print(f"raius: {r}, circumference: {c}")
````

Grade Processing

````python
decimals = np.array([85.7, 92.3, 78.9, 88.1, 95.6, 73.2])
print("Decimals:")
print(decimals)
````

Round to nearest integer

````python
rounded = np.round(decimals)
print("Rounded to nearest integer:")
print(rounded)
````

Clip to valid range [0, 100]

````python
clip = np.clip(decimals, 0, 100)
print("Clipped to [0, 100]:")
print(clip)
````

Calculate absolute difference from 85 (target)

````python
absolute = np.abs(decimals - 85)
print("Absolute difference from 85:")
print(absolute)
````

Find sign: above(+1) or below(-1) target

````python
sign = np.sign(decimals - 85)
print("Sign: above(+1) or below(-1) target:")
print(sign)
````
