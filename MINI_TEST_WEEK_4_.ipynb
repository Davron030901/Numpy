#Stack Operations

```python
month_1 = np.array([100, 120, 110, 130])
month_2 = np.array([105, 125, 115, 135])
month_3 = np.array([110, 130, 120, 140])

month = np.vstack((month_1, month_2, month_3))
id = np.array([1, 2, 3]).reshape(-1, 1)
print(month)
month_id = np.hstack((id, month))

print("3 months of sales:")
print(month_id)
```

#Split and Analyze

```python
revenue = np.array([1000, 1100, 1200, 1150, 1250, 1300,
           1350, 1400, 1450, 1500, 1550, 1600])

quarter = np.array(['Q1', 'Q2', 'Q3', 'Q4'])

quarter_revenue = np.split(revenue, 4)
quarter_revenue = np.array(quarter_revenue)
print(quarter_revenue)
avg_quarter = quarter_revenue.mean(axis=1)
print(f'quarter sum: {quarter_revenue.sum(axis=1)}')
print(f'quarter mean: {avg_quarter}')

print(f'Best {quarter[avg_quarter.argmax()]} quarter {avg_quarter[avg_quarter.argmax()]}')
```

#Reshape Challenge

```python
arr = np.arange(1, 25)
print('Original array:')
print(arr)

print('Reshape 2D array:')

r1 = arr.reshape(4, 6)
print("Array 4 rows, 6 columns")
print(r1)

r2 = arr.reshape(6, -1)
print("Array 6 rows, 3 columns")
print(r2)

r3 = arr.reshape(-1, 3)
print("Array 3 rows, 8 columns")
print(r3)

flattened = r1.flatten()
print("Flattened array:")
print(flattened)
print("Equal:", np.array_equal(arr, flattened))
```

#Flatten and Stats

```python
np.random.seed(42)
data = np.random.randint(50, 100, (5, 6))

flattened = data.flatten()
print("Flattened array:")
print(flattened)
```

```python
overall_min = flattened.min()
print("Min", overall_min)
overall_max = flattened.max()
print("Max", overall_max)
overall_mean = flattened.mean()
print("Mean", overall_mean)
overall_std = flattened.std()
print(f"Std: {overall_std:.2f}")
```

```python
each_student_avg = data.mean(axis=1)
print("Average of each student:")
print(each_student_avg)

each_subject_avg = data.mean(axis=0)
print("Average of each subject:")
print(each_subject_avg)
```

#Save and Load

```python
products = np.array(['Laptop', 'Phone', 'Tablet'], dtype=str)
prices = np.array([1200, 800, 500])
stock = np.array([50, 100, 75])
```

```python
saved = np.savez("inventory.npz", products=products, prices=prices, stock=stock)
print("Successful data")
```

```python
load = np.load("inventory.npz")
print(load.files)
```

```python
total = load['prices'] * load['stock']
print("Total:", total)
```

```python
summa = total.sum()
print("Sum:", summa)
```

```python
avg_price = total.mean()
print(f"Average price: {avg_price:.2f}")
```
