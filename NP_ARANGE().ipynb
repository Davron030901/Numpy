# NP.ARANGE()

Python range

```python
for i in range(5):
  print(i)
```

NumPy arange

```python
import numpy as np
arr=np.arange(5)
print(arr)
```

**Farqi:**
- `range()` = iterator (bir-bir)
- `np.arange()` = array (hammasi bir vaqtda)

**Real-world:**

Pixel koordinatalari:

```python
pixel_coords=np.arange(0,256)
print(pixel_coords)
```

Frame numbers:

```python
frame_numbers=np.arange(0,1000)
print(frame_numbers)
```

Index array:

```python
index_array=np.arange(100)
print(index_array)
```

0 to 10 (10 is not included!)

```python
arr0to10=np.arange(10)
print(arr1to10)
```

1 to 10

```python
arr1to10=np.arange(1,10)
print(arr1to10)
```

From 5 to 15

```python
arr5to15=np.arange(5,15)
print(arr5to15)
```

0 to 10, 2 steps

```python
arr0to10step2=np.arange(0,10,2)
print(arr0to10step2)
```

0 to 20, 5 steps

```python
arr0to20step5=np.arange(0,20,5)
print(arr0to20step5)
```

From 10 to 0, -1 step (reverse!)

```python
arr10to0step1=np.arange(10,0,-1)
print(arr10to0step1)
```

0 to 1, 0.1 step

```python
arr0to1step01=np.arange(0,1,0.1)
print(arr0to1step01)
```

0 to 5, 0.5 step


```python
arr0to5step05=np.arange(0,5,0.5)
print(arr0to5step05)
```

10 to 5, 1.1 step

```python
arr10to5step11=np.arange(10,5,-1.1)
print(arr10to5step11)
```

From 0 to 15

```python
arr20to0=np.arange(20,0,-1)
print(arr20to0)
```

0 to 2, 0.2 steps

```python
arroto2step02=np.arange(0,2,0.2)
print(arroto2step02)
```

Create a 2D array
0 to 11, then reshape to 3x4

```python
arr2D=np.arange(12).reshape(3,4)
print(arr2D)
```
