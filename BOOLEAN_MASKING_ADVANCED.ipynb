# Multiple Conditions (&, |, ~)! ðŸŽ­

````python
scores = np.array([55, 72, 88, 45, 91, 67, 95, 52, 78, 85])
print("Student scores:")
print(scores)
````

 AND condition (&)

````python
print("Students with scores between 70 and 90:")
mask = (scores >= 70) & (scores <= 90)
result = scores[mask]
print(f"Mask: {mask}")
print(f"Scores:{result}")
print(f"Count: {mask.sum()}")
````

OR condition (|)

````python
print("Students with score < 60 OR score > 90:")
mask = (scores < 60) | (scores > 90)
result = scores[mask]
print(f"Mask: {mask}")
print(f"Scores:{result}")
print(f"Count: {mask.sum()}")
````

NOT condition (~)

````python
print("Students who did NOT get exactly 72:")
mask = ~(scores == 72)
result = scores[mask]
print(f"Mask: {mask}")
print(f"Scores:{result}")
print(f"Count: {mask.sum()}")
````

np.logical_ Functions! ðŸ”§*

````python
arr = np.array([10, 20, 30, 40, 50, 60])
print("Array:")
print(arr)
````

np.logical_and

````python
print("np.logical_and (20 < x < 50):")
mask = np.logical_and(arr > 20, arr < 50)
result = arr[mask]
print(f"Mask: {mask}")
print(f"Values:{result}")
````

np.logical_or

````python
print("np.logical_or (x < 25 OR > 45)")
mask = np.logical_or(arr < 25, arr > 45)
result = arr[mask]
print(f"Mask: {mask}")
print(f"Values:{result}")
````

np.logical_not

````python
print("np.logical_not (x == 30):")
mask = np.logical_not(arr == 30)
result = arr[mask]
print(f"Mask: {mask}")
print(f"Values:{result}")
````

Multiple logical operations

````python
print("Complex: (x > 15 AND x < 45) OR x == 60:")
mask1 = np.logical_and(arr > 15, arr < 45)
mask2 = arr == 60
mask = np.logical_or(mask1, mask2)
result = arr[mask]
print(f"Mask1: {mask1}")
print(f"Mask2: {mask2}")
print(f"Mask: {mask}")
print(f"Values:{result}")
````

#np.isin() - Membership! ðŸ”

Student IDs

````python
all_students = np.array([101, 102, 103, 104, 105, 106, 107, 108])
print("All student IDs:")
print(all_students)
````

Present students

````python
present = [102, 104, 106, 108]
print(f"Present students: {present}")
````

Check who is present

````python
mask = np.isin(all_students, present)
present_ids = all_students[mask]
print(f"Present mask: {mask}")
print(f"Present IDs: {present_ids}")
print(f"Present count: {mask.sum()}")
````

Check who is absent (NOT present)

````python
absent_mask = ~np.isin(all_students, present)
absent_ids = all_students[absent_mask]
print(f"Absent mask: {absent_mask}")
print(f"Absent IDs: {absent_ids}")
print(f"Absent count: {absent_mask.sum()}")
````

#Boolean + Fancy Combo! ðŸš€

````python
products = np.array(['Laptop', 'Phone', 'Tablet', 'Monitor',
                     'Keyboard', 'Mouse', 'Headset'], dtype=str)
prices = np.array([1200, 800, 500, 300, 150, 50, 200])

print("Products and prices:")
for product, price in zip(products, prices):
  print(f"{product}: ${price}")
````

Find expensive items (price > 400)

````python
expensive_mask = prices > 400
print(f"Expensive mask: {expensive_mask}")
````

Get indices of expensive items

````python
expensive_indices = np.where(expensive_mask)[0]
print(f"Expensive indices: {expensive_indices}")
````

Select expensive products and prices

````python
expensive_products = products[expensive_mask]
expensive_prices = prices[expensive_mask]
print("Expensive items:")
for product, price in zip(expensive_products, expensive_prices):
  print(f"{product}: ${price}")
````

Apply 10% discount using indices

````python
discount_prices = prices.copy()
discount_prices[expensive_indices] = discount_prices[expensive_indices] * 0.9
print("After 10% discount on expensive items:")
for product, price in zip(products, discount_prices):
  print(f"{product}: ${price:.2f}")
````

#Temperature Filter

````python
temperature = np.array([18, 25, 22, 30, 28, 15, 20])
print("Weekly temperatures:")
print(temperature)
````

````python
mask = (temperature >= 20) & (temperature < 28)
comfortable = temperature[mask]
print(f"Mask: {mask}")
print(f"Comfortable temperatures:{comfortable}")
print(f"Count comfortable temperature: {mask.sum()}")
print(f"Average comfortable day: {comfortable.mean():.1f}")
````

````python
mask = (temperature < 20) | (temperature > 27)
uncomfortable = temperature[mask]
print(f"Mask: {mask}")
print(f"Uncomfortable temperatures:{uncomfortable}")
print(f"Count uncomfortable temperature: {mask.sum()}")
````

#Product Inventory

````python
products = np.array(['Apple', 'Banana', 'Orange', 'Grape', 'Mango'])
stocks = np.array([50, 0, 30, 0, 20])
prices = np.array([2.5, 1.5, 3.0, 4.0, 3.5])

print("Market")
for product, price, stock in zip(products, prices, stocks):
  print(f"{product}-- Price: ${price:.2f}, Stock: {stock}")
````

````python
stock = 0
mask = np.isin(stocks, stock)
empty_stocks = products[mask]
print(f"Mask: {mask}")
print(f"Empty stocks products:{empty_stocks}")
````

````python
mask = prices > 2.5
expensive = products[mask]
print(f"Mask: {mask}")
print(f"Expensive products:{expensive}")
````

````python
for product, price, stock in zip(expensive, prices[mask], stocks[mask]):
  print(f"{product}-- Price: ${price:.2f}, Stock: {stock}")
````
